```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("new_train_cat.csv")
train = data.drop(['trip_duration', 'log_trip_duration'], axis=1)
train.head()
```

```python
scaler = MinMaxScaler().fit(train) #Scaling data to 0-1 range
train_scaled = pd.DataFrame(scaler.transform(train))
train_scaled.head()
```

```python
def rmsle(real, predicted): # Caluculating rmsle score
    sum=0.0
    for x in range(len(predicted)):
        p = np.log(predicted[x]+1)
        r = np.log(real[x]+1)
        sum = sum + (p - r)**2
    return (sum/len(predicted))**0.5
```

```python
#splitting data
X_train, X_valid, y_train, y_valid = train_test_split(train_scaled, data['log_trip_duration'], test_size=0.2, random_state=43)
```

```python
#specifying network architecture
model = Sequential()
model.add(Dense(32, input_dim=65, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation ='relu'))
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
#training network
model.fit(X_train.values, y_train.values, batch_size=500, epochs=50, validation_data=(X_valid.values, y_valid.values), verbose=2)
```

```python
pred = np.abs(model.predict(X_valid.values, verbose=1)) #predicting on validation data
```

```python
print("RMSLE = ",rmsle(np.exp(y_valid.values), np.exp(pred))) #displaying cv scores
print("RMSE = ",np.sqrt(mean_squared_error(np.exp(y_valid),np.exp(pred))))
print("r2_score =",r2_score(np.exp(y_valid), np.exp(pred)))
```

```python
testd  = pd.read_csv("new_test_cat.csv") 
ids = pd.DataFrame(testd['id'])
test = testd.drop(['id'], axis=1)
test.head()
```

```python
test_scaled = pd.DataFrame(scaler.transform(test)) #scaling kaggle test data
test_scaled.head()
```

```python
predt = np.abs(model.predict(test_scaled.values, verbose=1))
```

```python
sub = pd.DataFrame(np.exp(predt), columns=['trip_duration'])
```

```python
submission = pd.concat([ids, sub], axis =1)
```
